## 1. Preparing the notebook

```python
# Import libraries
import pandas as pd
import numpy as np
import os
```

```python
#Importing data
ords_prods_merge = pd.read_pickle(r'C:\Users\ahamr\Desktop\12-2023 Instacart Basket Analysis\02 Data\Prepared Data\orders_products_merged_derived.pkl')
```

## 2. Aggregating Data with agg()

```python
# Grouping and aggregating the df 
ords_prods_merge.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

## 3. Analyzing the result.

As we can see, the maximum and minimum values are the same as in the subset because they were present in the subset, but the average values are different because the dataframe of a million rows did not include all records.

## 4. Aggregating Data with transform() and loc()

```python
# Creating a new column “max_order”
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(15)
```

```python
# Deriving Columns with loc()
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Checking the output
ords_prods_merge['loyalty_flag'].value_counts(dropna= False)
```

```python
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

## 5. Determining is whether the prices of products purchased by loyal customers differ from those purchased by regular or new customers.

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean']})
```

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['sum']})
```

## 6. Create a spending flag for each user based on the average price across all their orders using the criteria.

```python
# Creating a new column “avg_price”
ords_prods_merge['avg_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head(15)
```

```python
# Deriving the column with loc()
ords_prods_merge.loc[ords_prods_merge['avg_price'] < 10, 'spending flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_price'] >= 10, 'spending flag'] = 'High spender'
```

```python
# Checking the output
ords_prods_merge['spending flag'].value_counts(dropna= False)
```

```python
ords_prods_merge[['user_id', 'spending flag', 'avg_price']].head(10)
```

## 7. Creating an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column.

```python
# Creating a new column “median_order_frequency”
ords_prods_merge['median_order_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head(15)
```

```python
# Deriving the column with loc()
ords_prods_merge.loc[ords_prods_merge['median_order_frequency'] > 20, 'frequency flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_order_frequency'] > 10) & (ords_prods_merge['median_order_frequency'] <= 20), 'frequency flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_order_frequency'] <= 10, 'frequency flag'] = 'Frequent customer'
```

```python
# Checking the output
ords_prods_merge['frequency flag'].value_counts(dropna= False)
```

```python
ords_prods_merge[['user_id', 'frequency flag', 'median_order_frequency']].head(15)
```
